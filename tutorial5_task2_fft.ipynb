# Task 2: Signal Analysis using FFT

## Introduction

In this task, we analyze signals composed of multiple sine waves using the Fast Fourier Transform (FFT). We visualize the signals both in the time domain and frequency domain, and also observe the effects of adding noise.

---

## Import Libraries

```python
import os
import sys
from pathlib import Path
from matplotlib import pyplot as plt

# Ensure the repository root is on the path for module imports
sys.path.insert(0, str(Path().resolve().parents[1]))

from src.signal_generation import sum_sine_waves, add_noise
from src.fft_utils import compute_fft
```


## Generate and Plot Two-Sine Signal

A signal composed of two sine waves and visualize it in the time domain.

```python
fs = 1000          # Sampling frequency (Hz)
duration = 2.0     # Signal duration (seconds)
components = [(50, 1.0), (120, 0.5)]  # (frequency, amplitude) pairs

# Generate signal
t, signal = sum_sine_waves(components, fs, duration)

# Plot signal
plt.figure()
plt.plot(t, signal)
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.title("Two-sine signal")
plt.show()
```


## FFT Analysis of the Two-Sine Signal

Visualize the FFT to identify the frequency components of the signal.

```python
freqs, mag = compute_fft(signal, fs, full_range=True)

# Plot FFT magnitude spectrum
plt.figure()
plt.plot(freqs, mag)
plt.xlim(0, fs)
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.title("FFT")
plt.show()
```

## Generate and Plot Noisy Signal

Noise to the original signal and visualize the noisy signal in the time domain.

```python
# Add noise to signal
noisy_signal = add_noise(signal, 0.5)

# Plot noisy signal
plt.figure()
plt.plot(t, noisy_signal)
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.title("Noisy signal")
plt.show()
```

## FFT Analysis of the Noisy Signal

Compute and visualize the FFT of the noisy signal to see the impact of noise on frequency analysis.

```python
freqs_n, mag_n = compute_fft(noisy_signal, fs, full_range=True)

# Plot FFT magnitude spectrum of noisy signal
plt.figure()
plt.plot(freqs_n, mag_n)
plt.xlim(0, fs)
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.title("FFT of noisy signal")
plt.show()
```
